# The First Flares from TESS

This is a quick-and-dirty exploration of the first *public* "alert data" that TESS provided for 79 stars with possible transits. These are the flares that immediately stood out by eye, and should not be assumed to be a complete census of stellar activity from this sample.

- @jradavenport

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib
from astropy.table import Table
from scipy.optimize import curve_fit
from glob import glob
from lightkurve.targetpixelfile import TessTargetPixelFile

matplotlib.rcParams.update({'font.size':18})
matplotlib.rcParams.update({'font.family':'serif'})

ftype = '.pdf'
```

while looking in the other notebook (quick_plots) I manually found some flares. Let's zoom in on them and admire the bounty of flares (and superflares) that TESS will discover!

```python
files = glob('tess-data-alerts/*lc.fits')

k=0
data = Table.read(files[k], format='fits')
ok = data['QUALITY'] == 0

plt.figure(figsize=(10,5))
plt.plot(data['TIME'][ok], (data['PDCSAP_FLUX'][ok] / np.nanmedian(data['PDCSAP_FLUX'][ok])) - 1 , 
         '-o', label='PDCSAP FLUX', alpha=0.7)
plt.xlabel('BJD - 2457000 (days)')
plt.ylabel('Relative Flux')
plt.title(files[k][49:-5])
plt.xlim(1342.6, 1342.7)
plt.savefig(files[k][:-5]+'_flare1.png', dpi=300, bbox_inches='tight', pad_inches=0.25)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(data['TIME'][ok], (data['PDCSAP_FLUX'][ok] / np.nanmedian(data['PDCSAP_FLUX'][ok])) - 1 , 
         '-o', label='PDCSAP FLUX', alpha=0.7)
plt.xlabel('BJD - 2457000 (days)')
plt.ylabel('Relative Flux')
plt.title(files[k][49:-5])
plt.xlim(1348.1, 1348.5)
plt.savefig(files[k][:-5]+'_flare2.png', dpi=300, bbox_inches='tight', pad_inches=0.25)
plt.show()
```
